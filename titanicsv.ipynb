**Pretprocesiranje podataka - Brisanje nepotrebnih kolona i kreiranje novih osobina**

---

 1. Izbacivanje PassengerId, Ticket i Cabin kolona



```python
putnici = putnici.drop(['PassengerId', 'Ticket', 'Cabin'], axis=1)
```

```python
putnici
```

2. Kreiranje osobine Title na osnovu osobine Name

```python
putnici['Title']=putnici['Name'].str.split(",", expand=True)[1].str.split(".", expand=True)[0]
print(pd.crosstab(putnici['Title'], putnici['Sex']))
putnici=putnici.drop(['Name'], axis=1)
putnici
```

```python
putnici['Title']=putnici['Title'].replace([' Lady', ' the Countess', ' Capt', ' Col', ' Don',' Dr', ' Major', ' Rev', ' Sir', ' Jonkheer', ' Dona'], 'Rare')
putnici['Title'] = putnici['Title'].replace(' Mme', 'Mrs')
putnici['Title'] = putnici['Title'].replace([' Mlle', ' Ms'], 'Miss')
putnici['Title'] = putnici['Title'].replace(' Master', 'Master')
putnici['Title'] = putnici['Title'].replace(' Miss', 'Miss')
putnici['Title'] = putnici['Title'].replace(' Mr', 'Mr')
putnici['Title'] = putnici['Title'].replace(' Mrs', 'Mrs')
putnici[['Title', 'Survived']].groupby(['Title']).mean()
putnici[['Title', 'Age']].groupby(['Title']).mean()
pd.crosstab(putnici['Title'], putnici['Sex'])
```

Nove osobine FamilySize i Alone na osnovu SibSp i Parch

```python
putnici['FamilySize'] = putnici['SibSp'] + putnici['Parch'] + 1
putnici[['FamilySize', 'Survived']].groupby(['FamilySize']).mean().sort_values(by='Survived', ascending=False)
```

```python
putnici['Alone'] = 0
putnici.loc[putnici['FamilySize'] == 1, 'Alone'] = 1
putnici[['Alone', 'Survived']].groupby(['Alone']).mean()
putnici = putnici.drop(['SibSp', 'Parch'], axis=1)
putnici
```
