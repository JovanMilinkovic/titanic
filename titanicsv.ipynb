**Pretprocesiranje podataka - Brisanje nepotrebnih kolona i kreiranje novih osobina**

---

 1. Izbacivanje PassengerId, Ticket i Cabin kolona



```python
putnici = putnici.drop(['PassengerId', 'Ticket', 'Cabin'], axis=1)
```

```python
putnici
```

2. Kreiranje osobine Title na osnovu osobine Name

```python
putnici['Title']=putnici['Name'].str.split(",", expand=True)[1].str.split(".", expand=True)[0]
print(pd.crosstab(putnici['Title'], putnici['Sex']))
putnici=putnici.drop(['Name'], axis=1)
putnici
```

```python
putnici['Title']=putnici['Title'].replace([' Lady', ' the Countess', ' Capt', ' Col', ' Don',' Dr', ' Major', ' Rev', ' Sir', ' Jonkheer', ' Dona'], 'Rare')
putnici['Title'] = putnici['Title'].replace(' Mme', 'Mrs')
putnici['Title'] = putnici['Title'].replace([' Mlle', ' Ms'], 'Miss')
putnici['Title'] = putnici['Title'].replace(' Master', 'Master')
putnici['Title'] = putnici['Title'].replace(' Miss', 'Miss')
putnici['Title'] = putnici['Title'].replace(' Mr', 'Mr')
putnici['Title'] = putnici['Title'].replace(' Mrs', 'Mrs')
putnici[['Title', 'Survived']].groupby(['Title']).mean()
putnici[['Title', 'Age']].groupby(['Title']).mean()
pd.crosstab(putnici['Title'], putnici['Sex'])
```

3. Nove osobine FamilySize i Alone na osnovu SibSp i Parch

```python
putnici['FamilySize'] = putnici['SibSp'] + putnici['Parch'] + 1
putnici[['FamilySize', 'Survived']].groupby(['FamilySize']).mean().sort_values(by='Survived', ascending=False)
```

```python
putnici['Alone'] = 0
putnici.loc[putnici['FamilySize'] == 1, 'Alone'] = 1
putnici[['Alone', 'Survived']].groupby(['Alone']).mean()
putnici = putnici.drop(['SibSp', 'Parch'], axis=1)
putnici
```

4. Popunjavanje podataka koji nedostaju

```python
putnici[putnici['Embarked'].isna()]
```

```python
putnici['Embarked'] = putnici['Embarked'].fillna('S')
```

```python
print(putnici['Age'].mean())
putnici[['Title', 'Age']].groupby(['Title']).mean()
```

```python
putnici.loc[(putnici['Age'].isna()) & (putnici['Title'] == 'Master'), 'Age'] = 5
putnici.loc[(putnici['Age'].isna()) & (putnici['Title'] == 'Miss'), 'Age'] = 22
putnici.loc[(putnici['Age'].isna()) & (putnici['Title'] == 'Mr'), 'Age'] = 32
putnici.loc[(putnici['Age'].isna()) & (putnici['Title'] == 'Mrs'), 'Age'] = 36
putnici.loc[(putnici['Age'].isna()) & (putnici['Title'] == 'Rare'), 'Age'] = 46
```

**Konverzija osobina u odgovarajuce tipove podataka**

---
1. Konvertovanje osobina iz stringova u kategoricke numericke vrednosti


```python
putnici['Sex']=putnici['Sex'].map({'male': 0, 'female': 1})
putnici['Embarked']=putnici['Embarked'].map({"S": 0, "C": 1, "Q": 2})
putnici['Title']=putnici['Title'].map({'Mr': 1, 'Miss': 2, 'Mrs': 3, 'Master': 4, 'Rare': 5})
putnici.head(25)
```

2. Konvertovanje osobina iz kontinualnih numeriÄkih u kategoriÄke numeriÄke vrednosti

```python
pd.qcut(putnici['Fare'], 4) #moglo je da se skrati i da se umesto [43] koda napise: pd.qcut(putnici['Fare'], 4, labels=['0' ,'1', '2', '3'])
```

```python
putnici.loc[putnici['Fare']<=7.91, 'Fare'] = 0
putnici.loc[(putnici['Fare']>7.91) & (putnici['Fare']<=14.454), 'Fare'] = 1
putnici.loc[(putnici['Fare']>14.454) & (putnici['Fare']<= 31), 'Fare'] = 2
putnici.loc[putnici['Fare']>31] = 3
putnici['Fare']=putnici['Fare'].astype(int)
```

```python
print(putnici[putnici['Age'] == putnici['Age'].max()])
```

```python
putnici.loc[putnici['Age'] <= 16, 'Age'] = 0
putnici.loc[(putnici['Age'] > 16) & (putnici['Age'] <= 32), 'Age'] = 1
putnici.loc[(putnici['Age'] > 32) & (putnici['Age'] <= 48), 'Age'] = 2
putnici.loc[(putnici['Age'] > 48) & (putnici['Age'] <= 64), 'Age'] = 3
putnici.loc[putnici['Age'] > 64, 'Age'] = 4
putnici['Age'] = putnici['Age'].astype(int)
putnici
```

```python
from sklearn.linear_model import LogisticRegression	# LogistiÄka regresija
from sklearn.neighbors import KNeighborsClassifier	# K najbliÅ¾ih suseda
from sklearn.tree import DecisionTreeClassifier		# Stablo odluÄivanja
from sklearn.ensemble import RandomForestClassifier	# SluÄajna Å¡uma
from sklearn.svm import SVC				# Metoda potpornih vektora
from sklearn.model_selection import train_test_split		# podela podataka trening i test
from sklearn.metrics import accuracy_score			# merenje taÄnosti modela
```

```python
y = putnici['Survived']
X = putnici.drop('Survived', axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
#Logisticka regresija
model_1 = LogisticRegression(solver='lbfgs', max_iter=200)
model_1.fit(X_train, y_train)
predvidjanje=model_1.predict(X_test)
print('Logisticka regresija: ', accuracy_score(predvidjanje, y_test))

#K najbliÅ¾ih suseda
model_2 = KNeighborsClassifier(n_neighbors=3)
model_2.fit(X_train, y_train)
predvidjanje = model_2.predict(X_test)
print('K najblizih suseda: ', accuracy_score(predvidjanje, y_test))

#Stablo odluÄivanja
model_3 = DecisionTreeClassifier()
model_3.fit(X_train, y_train)
predvidjanje = model_3.predict(X_test)
print('Stablo odlucivanja: ', accuracy_score(predvidjanje, y_test))

#SluÄajna Å¡uma
model_4 = RandomForestClassifier(n_estimators=100)
model_4.fit(X_train, y_train)
predvidjanje = model_4.predict(X_test)
print('Slucajna suma: ', accuracy_score(predvidjanje, y_test))

#Metoda potpornih vektora
model_5 = SVC(gamma='scale')
model_5.fit(X_train, y_train)
predvidjanje = model_5.predict(X_test)
print('Metoda potpornih vektora: ', accuracy_score(predvidjanje, y_test))
```
